**One thing noticable is that `TotalCharges` should be `float64`**

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df.info()
```

- **As it converts the values into `float64` but also we have got some `NaN` values which have to remove**
    - **Before that we have to know if there is some mathamatical relation between `MonthlyCharges` and `TotalCharges`**
    - **Then we will fillup according to the results**

```python
plt.figure(figsize=(10, 10))
sns.scatterplot(data = df, x = df['MonthlyCharges'], y = df['TotalCharges'],hue="Contract")
plt.show()
```

**So by seeing the Graph we can conclude that there is no as such relation between `MonthlyCharges` and `TotalCharges`, So we can fillup those `NaN` values with some common techniques** 

- **Before that we have to perform some statistical opretion to get insides**
  - **Getting a brief statistical description about each numerical column**
